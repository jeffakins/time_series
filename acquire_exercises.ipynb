```python
def get_sales_data():
    '''A function to get the sales data from zgulde.net'''
    
    # Initial retrieval:
    base_url = 'https://python.zgulde.net'
    response = requests.get(base_url + '/api/v1/sales')
    data = response.json()
    df = pd.DataFrame(data['payload']['sales'])
    
    num_pages = data['payload']['max_page']
    
    for i in range(num_pages):
        response = requests.get(base_url + data['payload']['next_page'])
        data = response.json()
        df = pd.concat([df, pd.DataFrame(data['payload']['sales'])]).reset_index()
        
    return df
```

```python
sales_list = get_sales_data()
```

```python
def create_payload_df(url, max_pages, target):
    page_list = []

    for i in range(1, max_pages + 1):
        response = requests.get(url + "/api/v1/sales?page=" + str(i))
        data = response.json()
        page_items = data['payload'][target]
        page_list += page_items

    return pd.DataFrame(page_list)
```

```python
base_url = 'https://python.zgulde.net'

sales_list = create_payload_df(base_url, 183, 'sales')
```
